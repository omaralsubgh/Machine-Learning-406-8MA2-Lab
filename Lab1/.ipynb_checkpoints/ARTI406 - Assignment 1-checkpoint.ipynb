# Assignment 1: Exploratory Data Analysis (EDA)

EDA is the first and most important step in any Machine Learning project.
Before building models, we must understand:

- What does the data represent?
- Are there missing values?
- Are there outliers?
- What patterns exist?
- Which variables influence others?

If we do not understand the data, we cannot build a good model.

### Why EDA is Important

In real-world machine learning projects:

    1- 70–80% of the time is spent on understanding and cleaning data
    2- Only 20–30% is spent building models

**EDA helps us:**

- Detect errors
- Identify trends
- Discover relationships
- Make business decisions

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("Chocolate_Sales.csv")

df.head(-1)
```

The dataset contains 3,283 rows and 6 columns, representing chocolate sales transactions across multiple countries.

### Check Missing Values

```python
# Check all values
print(df.isna())
```

```python
df = df.drop_duplicates()
df.shape
```

```python
print(df.isna().sum())
```

The Date column contains a significant number of missing values.
One missing value exists in the Country column.
Other columns are complete.

### Check duplicate rows

```python
# checking duplicate rows
df.duplicated()[df.duplicated()==True]
```

### No. of rows and columns

```python
# finding number of rows and columns

print("Shape (rows, columns): ", df.shape,"\n")

print("number of rows: ", df.shape[0])
print("number of columns: ", df.shape[1])
```

### Data type of columns

```python
# viewing the data types of columns
df.dtypes
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Amount'] = df['Amount'].replace(r'[\$,]', '', regex=True)
df['Amount'] = pd.to_numeric(df['Amount'])

df.dtypes
```

```python
# Statistical summary
df.describe(include='all')
```

- The average revenue per transaction is approximately 6,028.
- The maximum sale reached over 26,170, indicating high-value transactions.
- The average boxes shipped per order is around 164.
- There is high variability in both revenue and shipment quantity.

### Descriptive summary Statistics

### Univariate Analysis

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Boxes Shipped'], bins=20)
plt.title("Distribution of Boxes Shipped")
plt.show()
```

- Shows how shipment sizes are distributed
- Right skew = many small shipments, few large ones

### Distribution of Revenue

```python
sns.histplot(df['Amount'], bins=20)
plt.title("Distribution of Revenue")
plt.xlabel("Revenue")
plt.ylabel("Frequency")
plt.show()
```

Revenue distribution shows variability with some large transactions.
The data appears right-skewed, indicating fewer high-value sales.


## Bivariate Analysis

### Revenue by Country

```python
country_revenue = df.groupby('Country')['Amount'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
country_revenue.plot(kind='bar')
plt.title("Total Revenue by Country")
plt.ylabel("Revenue")
plt.show()

country_revenue
```

Australia generates the highest total revenue among all countries.
This suggests strong market performance in that region.

### Revenue by Product

```python
product_revenue = df.groupby('Product')['Amount'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
product_revenue.plot(kind='bar')
plt.title("Revenue by Product")
plt.ylabel("Revenue")
plt.show()

product_revenue
```

Smooth Silky Salty is the top-performing product in terms of revenue.
Product performance varies significantly across the portfolio.


### Revenue by Sales Person

```python
salesperson_revenue = df.groupby('Sales Person')['Amount'].sum().sort_values(ascending=False)

salesperson_revenue.head(10)
```

Ches Bonnell is the highest revenue-generating sales person.
Overall revenue distribution among sales staff is relatively balanced.

### Boxes vs Revenue Relationship

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Boxes Shipped', y='Amount', data=df)
plt.title("Boxes Shipped vs Revenue")
plt.show()
```

### Correlation Matrix

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[['Boxes Shipped', 'Amount']].corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```

- Correlation close to 1 = strong positive relationship
- Close to 0 = weak relationship

The correlation between Boxes Shipped and Revenue is very weak.
This indicates that revenue is influenced more by product price than by quantity alone.

## Time-Based Analysis
### Monthly Revenue Trend

```python
df['Month'] = df['Date'].dt.to_period('M')

monthly_revenue = df.groupby('Month')['Amount'].sum()

plt.figure(figsize=(10,5))
monthly_revenue.plot()
plt.title("Monthly Revenue Trend")
plt.ylabel("Revenue")
plt.show()
```

Revenue fluctuates over time, indicating possible seasonal patterns.
However, missing date values limit the reliability of time-based analysis.

## Final Insights

- The dataset contains 3,283 chocolate sales transactions.
- Australia is the highest revenue-generating country.
- Smooth Silky Salty is the best-performing product.
- Ches Bonnell is the top sales performer.
- Revenue is not strongly correlated with the number of boxes shipped.
- Data contains missing date values that affect time-series reliability.

Overall, business performance varies by region and product category, with pricing playing a significant role in revenue generation.
